```python
import pandas as pd
from pandas import DataFrame
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
#retireve clean dataset
```

```python
df=(r'C:\Users\abrah\OneDrive\Desktop\Data Science projects\Video games sales\vgsales.csv')
pd.read_csv(df)
vsales=pd.read_csv(df)
vsales.shape
vsales.head()
vsales.columns
```

```python
#count null value
```

```python
print(vsales.isnull().sum())
```

```python
#remove null values
```

```python
vsales['Year'].fillna(vsales['Year'].mode()[0], inplace=True)

vsales['Publisher'].replace(np.nan, vsales['Publisher'].mode()[0], inplace=True)
```

```python
print(vsales.isnull().sum())
```

```python
#Remove values after 2016
```

```python
vsales = vsales[vsales.Year < 2017]
vsales.head()
```
